### First, let's import the generation function from our project

```python
from ficbot.character.name.generation import generate_name
```

### Next, let's set an (anime) image path

```python
img_path = "example/name/example.jpg"
```

```python
Image.open(img_path)
```

#### Generation function requires two things (besides the image itself): a model checkpoint, and a maps file. Let's start with the former.

```python
model_path = "example/name/simple_average.hdf5"
```

#### After loading, model checkpoint is already initialized and pretrained. The function will load the checkpoint on its own, so there is no need for any further action on our part.

```python
maps_path = "example/name/maps.pkl"
```

#### Name-generating models save maps automatically, before they start training. This is so that, when generating or resuming training from checkpoint, we knew which sequences were in the original corpus and / or could correctly import model outputs, using generated indices.

#### The final parameter we need is diversity. It is important to know, that the model does NOT generate a single character or an index. Instead it assignes each of the possible continuation tokens a probability it should be the one to use next.

```python
generate_name(img_path,
              model_path,
              maps_path,
              diversity=0.1)
```

#### If we always use only the most probable tokens, the model outputs will (duh!) be very repetitive. Quite boring, don't you think?

#### Thus, our sampling function takes a **diversity** parameter, which allows the model to be more "adventurous" in its choices. The higher our diversity is, the more likely our sampling function to take a less obvious token. 

#### The default diversity is 1.0, but feel free to play around with it!

### Now, it is finally time to generate our name!

```python
Image.open(img_path)
```

```python
generate_name(img_path,
              model_path,
              maps_path,
              diversity=1.0)
```

```python
generate_name(img_path,
              model_path,
              maps_path,
              diversity=1.2)
```

```python
generate_name(img_path,
              model_path,
              maps_path,
              diversity=2.0)
```

### Tip: the generated name is not a be-all end-all prediction for the image. If you execute the cell above again, you are very likely to get a different result!
