```python
mal_wimages = mal[mal.img_index != "-1"]
```

```python
mal_wimages.eng_name = mal_wimages.eng_name.map(lambda x: "@" + x + "$")
```

```python
mal_wimages[["eng_name", "img_index"]].head()
```

```python
import pickle
```

```python
with open("../models/name_generation/tf/checkpoints/maps.pkl", "rb") as nm:
    maps = pickle.load(nm)
```

```python
char_counts = maps['char_idx']
```

```python
for char in char_counts.keys():
    char_counts[char] = 0
    for name in mal_wimages["eng_name"]:
        char_counts[char] += name.count(char)
```

```python
char_counts
```

```python
rare_chars = {char: count for char, count in char_counts.items() if count < 100}
```

```python
rare_chars_set = set(rare_chars)
```
