```python
def create_document_text(row):
    """
    row: A dictionary or pandas Series containing:
      - char_name
      - char_bio
      - anime_title
      - anime_genres
      - anime_themes
      - anime_synopsis
      - anime_demographic (optional)
    """
    text_parts = [
        f"Character Name: {row['eng_name']}",
        f"Character Bio: {row['bio']}",
        f"Anime Title: {row['anime_title']}",
        f"Genres: {row['anime_genres'].replace('|', ', ')}",
        f"Themes: {row['anime_themes'].replace('|', ', ')}",
        f"Synopsis: {row['anime_synopsis']}"
    ]
    # Demographic is optional
    if 'anime_demographic' in row and row['anime_demographic']:
        text_parts.append(f"Demographic: {row['anime_demographic']}")

    # Join into one big text chunk
    return "\n".join(text_parts)
```

```python
print(create_document_text(top1000_dataset.iloc[0]))
```
